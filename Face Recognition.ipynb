# Face Recognition

```python
from matplotlib import pyplot as plt
import cv2
```

```python
#load image
image = cv2.imread("/home/mahsa/Downloads/test.jpg")
image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
#show image
plt.imshow(image)
```

```python
#Load the model
model = cv2.CascadeClassifier("/home/mahsa/Downloads/model.xml")
face = model.detectMultiScale(image)
```

```python
print(face)
```

```python
image.shape
```

```python
#Print the coordinates of points in pairs on each line
for row in face:
    for i in range(0, len(row), 2):
        print(row[i], row[i+1])
    
```

```python
#Draw rectangles
for rect in face:
   image =  cv2.rectangle(image, (rect[0], rect[1]), (rect[0] + rect[2], rect[1] + rect[3]), (255, 0, 0), 3)
```

```python
#Display rectangles in picture
plt.imshow(image)
```
